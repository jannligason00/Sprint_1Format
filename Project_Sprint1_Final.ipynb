#                        Correlation Analysis

### **Games Won vs Run Scored**

```python
df2.plot(kind='scatter', 
       x='Games Won',
       y='Runs Scored',
       title='Offense Relative to Winning')
plt.show()
```

The graph shows a high positive correlation of Runs scored to winning.

### **Games Won vs Runs Against**

```python
df2.plot(kind='scatter', 
       x='Games Won',
       y='Runs Against',
       title='Defense Relative to Winning')
plt.show()
```

Although it is not very clear, the graph indicates that there is a negative correlation between winning and Runs against.

### **Games Won vs (Runs Scored - Runs Against)**

```python
df2.plot(kind='scatter', 
       x='Games Won',
       y='Result',
       title='Difference of Offense to Defense Relative to Winning')
plt.show()
```

By subtracting the two strategy measurements, the graph shows a stronger and cleaner positive linear relation of the data. We can identify with this graph that we need a combination of offensive and defensive skill to win more games.

### **Games Lost vs Runs Scored**

```python
df2.plot(kind='scatter', 
       x='Games Lost',
       y='Runs Scored',
       title='Offense Relative to Losing')
plt.show()
```

The data points on the graph is all over the place, it does'nt provide a clear correlation.

### **Games Lost vs Runs Against**

```python
df2.plot(kind='scatter', 
       x='Games Lost',
       y='Runs Against',
       title='Defense Relative to Losing')
plt.show()
```

It shows that most data point are concentrated on the center with just a little of linear indication.

### **Games Lost vs (Runs Scored - Runs Against)**

```python
df2.plot(kind='scatter', 
       x='Games Lost',
       y='Result',
       title='Difference of Offense to Defense Relative to Losing')
plt.show()
```

This graph shows that there is a stong negative correlation of the difference. Offense alone cannot provide a positive output on the games. If a team had negative result in this correlation they have to double or triple their effort in defense to minimize their lost.
